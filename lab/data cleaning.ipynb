```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('layoff_data.csv')
```

```python
data.head()
```

```python
def replace_nan_with_nan(label: str):
    label_str = str(label).lower().strip()
    if label_str=='nan' or label_str=='unclear':
        return np.nan
    return label
```

```python
for col in data.columns:
    data[col] = data[col].apply(replace_nan_with_nan)
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data = data[data['No. of reported layoffs'].notna()]
```

```python
data.head()
```

```python
data.shape
```

```python
data.rename(columns={
    'Company (click name for more info)': 'company',
    'No. of reported layoffs': 'layoffs',
    'Announced/reported date': 'report_date',
    'Company Status': 'company_status'
}, inplace=True)
data.drop([
    '% of company laid off',
    'Source',
    'Notes'  
], axis=1, inplace=True)
```

```python
data.head()
```

```python
data['layoffs'] = data['layoffs'].apply(lambda val: int(str(val).replace(',', '')))
```

```python
data.to_csv('clean_layoffs.csv', index=False)
```
